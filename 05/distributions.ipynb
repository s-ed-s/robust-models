```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, expon, poisson, binom
```

```python
def show_func(title, x, y, y_dist):    
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.plot(x, y, color='blue')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{title}\nПлотность')
    plt.grid()
    
    plt.subplot(1, 2, 2)
    plt.plot(x, y_dist, color='blue')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{title}\nФункция распределения')
    plt.grid()
    
    plt.show()
```

```python
distributions = [
    [uniform, 'Равномерное распределение', {
        'loc' : 1, 'scale': 5
    }],
    [norm, 'Нормальное распределение', {
        'loc' : 1, 'scale': 3
    }],
    [expon, 'Экспоненцальное распределение', {
        'loc' : 1, 'scale': 5
    }],
    [poisson, 'Распределение Пуассона', {
        'mu' : 0.8
    }],
    [binom, 'Биномиальное распределение', {
        'n' : 10, 'p': 0.6
    }]    
]
```

```python
x = np.linspace(-10, 10, 1000)
for distribution, name, params in distributions:
    y = 0
    if distribution in [binom, poisson]:
        y = distribution.pmf(x, **params)
    else:
        y = distribution.pdf(x, **params)
    y_dist = distribution.cdf(x, **params)
    show_func(name, x, y, y_dist)
```
