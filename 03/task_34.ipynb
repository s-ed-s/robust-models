```python
import matplotlib.pyplot as plt
from scipy.misc import derivative
from math import sin
import numpy as np
```

```python
def func(x):
    return sin(x)
def derivative_func(x):
    return derivative(func, x, dx=1e-4)
```

```python
x = [x/100 for x in range(-500, 500, 1)]
y = [func(x) for x in x]
y_d = [derivative_func(x) for x in x]
```

```python
plt.plot(x, y, label='y=sin(x)')
plt.plot(x, y_d, label='y=df(x)/dx')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
plt.grid(True)

plt.show()
```

# task 4

```python
def f(x, y):
    return x**2-y**2
```

```python
x = np.linspace(-10, 10, 16)
y = np.linspace(-10, 10, 16)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)
```

#### Поверхность f (x ; y ) = x^2 − y^2

```python
ax = plt.axes(projection="3d")
ax.contour3D(X, Y, Z, 50, cmap="binary")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.title('f(x,y)=x^2-y^2')
plt.show()
```

#### Поле направлений f(x, y) = x^2 − y^2

```python
dx = np.ones(Z.shape)

plt.quiver(X, Y, dx, Z, color="green")
plt.title("Поле направлений f(x, y) = x^2 − y^2")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
